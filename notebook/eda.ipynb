```python
df.groupby('Interest_Rate')['Loan_Amount_Requested'].agg(['count', 'mean', 'median', 'min', 'max'])
```

```python
from scipy.stats import f_oneway

group1 = df[df['Interest_Rate'] == 1]['Loan_Amount_Requested']
group2 = df[df['Interest_Rate'] == 2]['Loan_Amount_Requested']
group3 = df[df['Interest_Rate'] == 3]['Loan_Amount_Requested']

f_stat, p_value = f_oneway(group1, group2, group3)

print("F-statistic:", f_stat)
print("p-value:", p_value)

if p_value > 0.05:
    print("p-value > 0.05: Fail to reject the null hypothesis. No significant difference in loan amount requested among interest rate categories.")
else:
    print("p-value ≤ 0.05: Reject the null hypothesis. There is a significant difference in loan amount requested among interest rate categories.")
```

📌 Key Insights

💸 Higher loan amounts are associated with higher interest rates.

Borrowers requesting larger loans are more likely to fall into category 3 (high interest).

⚠️ Risk-based pricing might be in play — larger loans may indicate higher credit risk, thus attracting higher rates.

📊 Distribution Spread Matters — category 3 has more variability, suggesting greater inconsistency in borrower profiles or loan purposes.

📝 Conclusion Summary

There is a statistically significant and practically meaningful difference in the loan amounts requested across different interest rate categories. Specifically, higher interest rate categories are associated with higher average loan requests.

```python
import pandas as pd
from scipy.stats import chi2_contingency

ct = pd.crosstab(df['Home_Owner'], df['Interest_Rate'])

chi2, p, dof, expected = chi2_contingency(ct)

print(f"Chi-square statistic: {chi2}")
print(f"Degrees of freedom: {dof}")
print(f"P-value: {p:.4f}")

alpha = 0.05  

if p < alpha:
    print("Result: Reject the null hypothesis.")
    print("Conclusion: There is a statistically significant association between 'House_Owner' and 'Interest_Rate'.")
else:
    print("Result: Fail to reject the null hypothesis.")
    print("Conclusion: There is no statistically significant association between 'House_Owner' and 'Interest_Rate'.")
```

| **Rent**            | Second-highest volume, similar pattern to Mortgage.                                |
| **Own**             | Very small group, but still skewed toward Interest Rate 2.                         |
| **Other**           | Negligible numbers; not influential in the analysis.                               |
| **Interest Rate 2** | Most common across all house ownership categories.                                 |


📝 Final Summary Conclusion

There is a strong and statistically significant association between house ownership status and interest rate levels. Specifically:

People with a mortgage or who rent are far more likely to fall into Interest Rate 2 or 3 categories.

Homeowners (Own) without a mortgage tend to appear less often and are relatively more stable, but still most commonly experience Interest Rate 2.

The "Other" category is statistically insignificant due to very small counts.
